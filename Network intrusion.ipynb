```python
# By observation we know that there are some catagorical features are present in the data set.
#We will convert those feature to numeric by one-hot-encoding.Pandas provide a vary beautiful function called .get_dummies() which will do all our work.
#First we will select which columns are catagorical by making the list of those column names.
cat_col=['protocol_type', 'service', 'flag']
#now we will get the dummy values of these columns and assign those to the new variable called dum_train/dum_test.
dum_train=pd.get_dummies(train_data[cat_col],drop_first=True)
dum_test=pd.get_dummies(test_data[cat_col],drop_first=True)
dum_train.info()
```

```python
# Now we will drop the previous catagorical variable from the dataset
train_data.drop(cat_col,axis=1,inplace=True)
test_data.drop(cat_col,axis=1,inplace=True)
```

```python
#In our train data we have one extra variable i.e target variable
#we will convert that using Encoder
from sklearn.preprocessing import LabelEncoder 
le= LabelEncoder()
train_data["class"]=le.fit_transform(train_data["class"])
train_data.info()
```

```python
#Now we will concatinate the our data and converted data
final_train=pd.concat((train_data,dum_train),axis=1)
final_test=pd.concat((test_data,dum_test),axis=1)
```

# Feature Selection by Ensamble Method:

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
```

```python
F_rf=RandomForestClassifier()
```

```python
F_rf.fit(final_train.drop("class",axis=1),final_train["class"])
```

```python
#Ploting the importance of the data.
feature_importance=F_rf.feature_importances_
#make importance relative to max importance
feature_importance=100.0*(feature_importance/ feature_importance.max()) # Standardizing the data
sorted_idx = np.argsort(feature_importance)
pos=np.arange(sorted_idx.shape[0])+.5
plt.figure(figsize=(30,30))
plt.subplot(1,2,2)
plt.barh(pos,feature_importance[sorted_idx],align="center")
plt.yticks(pos,final_train.drop("class",axis=1).columns[sorted_idx])
plt.xlabel("Relative importance")
plt.title("variable importance")
plt.show()
```

```python
#selecting the important features using the SelectFromModel() 
sklearn.feature_selection import SelectFromModel
sel=SelectFromModel(RandomForestClassifier(n_estimators=25))
```

```python
sel.fit(final_train.drop("class",axis=True),final_train["class"])
```

```python
len(sel.get_support())
```

```python
#splitting oyr the data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(final_train.drop("class",axis=1).iloc[:,sel.get_support()],final_train['class'],test_size=.3,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#Fitting the model
log_model=LogisticRegression()
```

```python
log_model.fit(x_train,y_train)
```

```python
#prediction
predict_train=log_model.predict(x_train)
predict_test=log_model.predict(x_test)
```

```python
from sklearn.metrics import roc_curve,classification_report,accuracy_score,auc,confusion_matrix
```

```python
print("Accuracy:",accuracy_score(y_train,predict_train),"\n","Classification report:\n",classification_report(y_train,predict_train))
print("Accuracy:",accuracy_score(y_test,predict_test),"\n","Classification report:\n",classification_report(y_test,predict_test))
```

```python
cm=confusion_matrix(y_test,predict_test)
#cploting confussion matrix
sns.heatmap(pd.DataFrame(cm),annot=True)
```

```python
#Determine the false position and true possitive rates
FPR, TPR, _=roc_curve(y_train,predict_train)
FPR1,TPR1,_1=roc_curve(y_test,predict_test)
```

```python
# Calculate the AUC
roc_auc=auc(FPR,TPR)
print("ROC AUC:%0.3f" % roc_auc)


roc_auc1=auc(FPR1,TPR1)
print("ROC AUC:%0.3f" % roc_auc1)
```

```python
#plot of a ROC curve
plt.figure(figsize=(10,10))
plt.plot(FPR,TPR,label="ROC curve (area=%0.3f)" % roc_auc)
plt.plot(FPR1,TPR1,label="ROC curve (area=%0.3f)" % roc_auc1)
plt.plot([0,1],[0,1],'k--')
plt.xlim([-0.05,1.0])
plt.xlim([0.0 , 1.05])
plt.xlabel("false possitive rate")
plt.ylabel("true possitive rate")
plt.title("ROC CURVE")
plt.legend(loc="lower right")
plt.show()
```
